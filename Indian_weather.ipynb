```python
X = dataset.drop(columns=['condition_text'],axis=1)
Y = dataset['condition_text']
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
X_train.shape,X_test.shape
```

```python
#analysis
from sklearn.feature_selection import mutual_info_classif
classifier = mutual_info_classif(X_train,Y_train)
classifier = pd.Series(classifier)
classifier.index = X_train.columns
classifier.sort_values(ascending=False).plot.bar(figsize=(12,8))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
tree = ExtraTreesClassifier()
tree.fit(X_train,Y_train)
feat_imp = pd.Series(tree.feature_importances_,index=X_train.columns)
feat_imp.nlargest(20).plot.bar(figsize=(12,10))
```

```python
#model selection
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
models = [RandomForestClassifier(max_features=0.5,max_samples=0.3)]
for model in models:
  model.fit(X_train,Y_train)
  X_train_prediction = model.predict(X_train)
  from sklearn.metrics import accuracy_score
  X_train_accuracy = accuracy_score(X_train_prediction,Y_train)
  print('Accuracy of the:',model,'=',X_train_accuracy)
  print('++++++++++++++++++++++++++++++++++++++++++++++++++')
```

```python
#test data
for model in models:
  X_test_prediction = model.predict(X_test)
  from sklearn.metrics import accuracy_score
  X_test_accuracy = accuracy_score(X_test_prediction,Y_test)
  print('Accuracy of the:',model,'=',X_test_accuracy)
```

```python
#confusion matrix
plt.figure(figsize=(12,10))
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(X_test_prediction,Y_test)
sns.heatmap(matrix,annot=True)
plt.show()
```
